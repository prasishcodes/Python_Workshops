```python
titanic = pd.read_csv("/content/drive/MyDrive/Python_Workshops/Copy of Titanic-Dataset.csv")
```

```python
subset = titanic[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
subset_p1 = subset[subset['Pclass'] == 1]
```

```python
mean_fare = subset_p1['Fare'].mean()
median_fare = subset_p1['Fare'].median()
max_fare = subset_p1['Fare'].max()
min_fare = subset_p1['Fare'].min()

mean_fare, median_fare, max_fare, min_fare
```

```python
embarked_dummies = pd.get_dummies(titanic['Embarked'], prefix='Embarked')
```

```python
titanic = pd.concat([titanic, embarked_dummies], axis=1)
```

```python
titanic = titanic.drop(columns=['Embarked'])
```

Problem No 4

```python
import matplotlib.pyplot as plt

survival_by_sex = titanic.groupby('Sex')['Survived'].mean()
print(survival_by_sex)

survival_by_sex.plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
titanic2 = pd.read_csv("/content/drive/MyDrive/Python_Workshops/Copy of Titanic-Dataset.csv")

grouped = titanic2.groupby(['Sex', 'Embarked'])['Survived'].mean()
print(grouped)

grouped.unstack().plot(kind='bar')
plt.title("Survival Rate by Gender and Embarkation Port")
plt.ylabel("Survival Rate")
plt.show()
```

```python
titanic['AgeGroup'] = pd.qcut(titanic['Age'], 5)
```

```python
survival_table = titanic.groupby(['Pclass', 'AgeGroup'])['Survived'].mean()
print(survival_table)
```

```python
survival_table.unstack().plot(kind="bar", figsize=(10,5))
plt.title("Survival Rate by Age Group and Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```
