```python
# Daily time (in hours): [study, entertainment, sleep]
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
```

```python
# Create empty lists for study time classifications
low_study = []      # less than 3 hours
moderate_study = [] # between 3 and 5 hours
high_study = []     # more than 5 hours

# Iterate over time_data and classify study hours
for day in time_data:
    study_hours = day[0]

    if study_hours < 3:
        low_study.append(study_hours)
    elif 3 <= study_hours <= 5:
        moderate_study.append(study_hours)
    else:  # study_hours > 5
        high_study.append(study_hours)

# Print the lists to verify
print("Low study time (< 3 hours):", low_study)
print("Moderate study time (3-5 hours):", moderate_study)
print("High study time (> 5 hours):", high_study)
```

```python
# 1. How many days had low study time?
low_count = len(low_study)
print(f"Days with low study time: {low_count}")

# 2. How many days had moderate study time?
moderate_count = len(moderate_study)
print(f"Days with moderate study time: {moderate_count}")

# 3. How many days had high study time?
high_count = len(high_study)
print(f"Days with high study time: {high_count}")
```

```python
# Convert study hours to minutes
study_minutes = []

for day in time_data:
    study_hours = day[0]
    study_minutes_value = study_hours * 60
    study_minutes.append(study_minutes_value)

# Print the converted values
print("Study hours converted to minutes:", study_minutes)
```

```python
# Create empty lists
study_hours = []
entertainment_hours = []
sleep_hours = []

# Extract values into each list
for day in time_data:
    study_hours.append(day[0])
    entertainment_hours.append(day[1])
    sleep_hours.append(day[2])

# Calculate averages
avg_study = sum(study_hours) / len(study_hours)
avg_entertainment = sum(entertainment_hours) / len(entertainment_hours)
avg_sleep = sum(sleep_hours) / len(sleep_hours)

# Print results
print(f"Average study hours: {avg_study:.2f}")
print(f"Average entertainment hours: {avg_entertainment:.2f}")
print(f"Average sleep hours: {avg_sleep:.2f}")
```

```python
import matplotlib.pyplot as plt

# Extract study and sleep hours
study = [day[0] for day in time_data]
sleep = [day[2] for day in time_data]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(study, sleep, color='blue', alpha=0.7)
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.title('Study vs Sleep Pattern')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.

    Args:
        nested_list (list): A list that may contain integers or other lists of integers.

    Returns:
        int: The total sum of all integers in the nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

# Test the function
test_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(test_list)
print(f"Sum of nested list: {result}")  # Output: 36
```

```python
def generate_permutations(s):
    """
    Generate all unique permutations of a string.

    Args:
        s (str): Input string

    Returns:
        list: List of all unique permutations
    """
    # Base case
    if len(s) <= 1:
        return [s]

    permutations = []
    for i in range(len(s)):
        # Extract current character
        current_char = s[i]

        # Remaining string
        remaining_chars = s[:i] + s[i+1:]

        # Generate permutations for remaining characters
        for perm in generate_permutations(remaining_chars):
            permutations.append(current_char + perm)

    return permutations

# Test the function
test_string = "abc"
permutations = generate_permutations(test_string)
print(f"Permutations of '{test_string}': {permutations}")
```

```python
def calculate_directory_size(directory):
    """
    Calculate the total size of a directory including all nested files and subdirectories.

    Args:
        directory (dict): Dictionary representing directory structure

    Returns:
        int: Total size in KB
    """
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
            # It's a subdirectory, recursively calculate its size
            total_size += calculate_directory_size(value)
        else:
            # It's a file, add its size
            total_size += value

    return total_size

# Test the function
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

total_size = calculate_directory_size(directory_structure)
print(f"Total directory size: {total_size} KB")  # Output: 1400 KB
```

```python
def min_coins(coins, amount):
    """
    Find the minimum number of coins needed to make up a given amount.

    Args:
        coins (list): List of coin denominations
        amount (int): Target amount

    Returns:
        int: Minimum number of coins needed, or -1 if not possible
    """
    # Initialize DP array with infinity
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed for amount 0

    # Fill DP array
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test the function
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"Minimum coins needed for amount {amount}: {result}")  # Output: 3
```

```python
def longest_common_subsequence(s1, s2):
    """
    Find the length of the longest common subsequence between two strings.

    Args:
        s1 (str): First string
        s2 (str): Second string

    Returns:
        int: Length of the longest common subsequence
    """
    m, n = len(s1), len(s2)

    # Create DP table
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

# Test the function
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"Length of LCS between '{s1}' and '{s2}': {result}")  # Output: 3
```

```python
def knapsack(weights, values, capacity):
    """
    Solve the 0/1 knapsack problem using dynamic programming.

    Args:
        weights (list): List of item weights
        values (list): List of item values
        capacity (int): Maximum weight capacity

    Returns:
        int: Maximum value that can be achieved
    """
    n = len(weights)

    # Create DP table
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                # Either include the item or don't
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                # Can't include the item
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

# Test the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"Maximum value for capacity {capacity}: {result}")  # Output: 9
```
